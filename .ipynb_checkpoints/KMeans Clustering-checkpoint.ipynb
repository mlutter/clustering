# KMeans Clustering

This is a demonstration of the kmeans clustering done against the default Iris dataset. We know initially that there will be three clusters representing the three species: Setosa, Versicolor and Virginica. Here we demonstrate how we might discover that three was the appropriate number by plotting the inertia of different cluster quantities and watching for the point where additional clusters offer only minimal gains. 

The kmeans algorithm works by positioning a 'centroid' of n_clusters in the data field, assigning each data point to it's nearest centroid then repositioning the centroid to the center of the cluster. Once the centroid moves, it reassigns the data points. If one of them changes clusters the algorithm repeats, if none change the algorithm has completed. 

```python
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#load the data
iris = datasets.load_iris() #load the dataset from sklearn
samples = iris['data']
ks = range(1, 6)
inertias = []

for k in ks:
    # Create a KMeans instance with k clusters: model
    # use init='k-means++' as parameter to avoid the random initialization problem. 
    model=KMeans(n_clusters=k, init='k-means++')
    
    # Fit model to samples
    model.fit(samples)
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
# Plot ks vs inertias to observe the appropriate number of clusters to choose. 
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()

#then, from the graph we decide three should be a good number of clusters, 
#which matches the reality of the three iris species. 
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(samples[:, [0,2]])


# Visualising the clusters (only two features shown), 
#leaving the cluster names generic since we presume not to know there were three species
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of Iris')
plt.xlabel('Sepal length (cm)')
plt.ylabel('Petal length (cm)')
plt.legend()
plt.show()
```
